```python

#  Explore the data and list down unique values in each feature
unique_values_info = {}
for column in data.columns:
    unique_values = data[column].unique()
    unique_values_info[column] = {'unique_values': unique_values, 'length': len(unique_values)}
```

### Data Cleaning



```python
# Find duplicate rows
duplicate_rows = data[data.duplicated()]

# Display the duplicate rows (optional)
print("Duplicate Rows:")
print(duplicate_rows)

# Drop duplicate rows and keep the first occurrence
data.drop_duplicates(keep='first', inplace=True)

# Display the DataFrame after removing duplicates
print("\nDataFrame after removing duplicates:")
```

```python
data
```

```python

z_scores = np.abs(stats.zscore(data.select_dtypes(include=np.number)))
outlier_indices = np.where(z_scores > 3)
```

```python
data = data.drop(data.index[outlier_indices[0]])
data
```

### Data Analysis


```python
# Step 1: Filter the data with age > 40 and salary < 5000
filtered_data = data[(data['Age'] > 40) & (data['Salary'] < 5000)]
filtered_data
```

```python
# Step 2: Plot the chart with age and salary
plt.scatter(filtered_data['Age'], filtered_data['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()
```

```python
# Step 3: Count the number of people from each place and represent it visually
place_counts = filtered_data['Place'].value_counts()
place_counts
```

```python
# Plotting the count of people from each place
place_counts.plot(kind='bar')
plt.xlabel('Place')
plt.ylabel('Number of People')
plt.title('Number of People from Each Place')
plt.show()
```

```python
# Perform one-hot encoding on the 'Company', 'Place', and 'Gender' columns
data_encoded = pd.get_dummies(data, columns=['Company', 'Place', 'Gender'])

# Display the encoded DataFrame
print(data_encoded.head())
```

```python
# Label encode 'Company'
label_encoder = LabelEncoder()
data['Company_encoded'] = label_encoder.fit_transform(data['Company'])

# Display the DataFrame with the label-encoded 'Company' column
print(data)
```
